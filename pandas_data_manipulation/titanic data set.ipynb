```python
# Step 10(a): Find the survival percentage of passengers in group "C"

cherbourg_total = len(titanic_data[titanic_data["Embarked"] == "C"])
cherbourg_survival = len(titanic_data[(titanic_data["Embarked"] == "C") & (titanic_data["Survived"] == 1)])

cherbourg_survival/cherbourg_total*100
```

```python
# Step 10(b): Find the survival percentage of passengers in group "Q"

queenstown_total = len(titanic_data[titanic_data["Embarked"] == "Q"])
queenstown_survival = len(titanic_data[(titanic_data["Embarked"] == "Q") & (titanic_data["Survived"] == 1)])

queenstown_survival/queenstown_total*100
```

```python
# Step 10(c): Find the survival percentage of passengers in group "S"

south_survival_total = len(titanic_data[titanic_data["Embarked"] == "S"])
south_survival_survival = len(titanic_data[(titanic_data["Embarked"] == "S") & (titanic_data["Survived"] == 1)])

south_survival_survival/south_survival_total*100
```
